```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/QSS20_panel_clean.csv')

region_mean = df.groupby(['Region', 'year'])['elec_access'].mean().unstack('Region')
region_mean.plot(figsize=(10, 5), marker = 'o', markersize = 3)
plt.title('Mean Electricity Access by Region, 2000-2022')
plt.xlabel('Year')
plt.ylabel('% of population with electricity')
plt.tight_layout()
plt.legend(loc = 'lower right', fontsize = 8)
plt.savefig('figgure 1.png', dpi=150)
plt.show()
```

```python
clean = df.dropna(subset=['elec_access', 'ind_employment'])
plt.figure(figsize=(10, 5))
plt.scatter(clean['elec_access'], clean['ind_employment'], alpha=0.3, s=15)
plt.title('Electricity Access vs Industrial Employment')
plt.xlabel('Electricity Access')
plt.ylabel('Industrial Employment')
plt.tight_layout()
plt.savefig('figure 2.png', dpi=150)
plt.show()
```
